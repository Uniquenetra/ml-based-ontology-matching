```python
from deeponto.align.mapping import ReferenceMapping
refs = ReferenceMapping.read_table_mappings("/content/drive/MyDrive/Thesis/OMIM2ORDO/full.tsv")
train_val = ReferenceMapping.read_table_mappings("/content/drive/MyDrive/Thesis/OMIM2ORDO/train+val.tsv")
```

```python
#BLBL, 0.6
from deeponto.align.mapping import EntityMapping
ents = EntityMapping.read_table_mappings("/content/drive/MyDrive/Thesis/OMIM2ORDO/withBLBL.csv", 0.6)

from deeponto.align.evaluation import AlignmentEvaluator
prec = AlignmentEvaluator.precision(ents, refs)
recl = AlignmentEvaluator.recall(ents, refs)
f1 = AlignmentEvaluator.f1(ents, refs, train_val) #only on test

prec, recl, f1
```
